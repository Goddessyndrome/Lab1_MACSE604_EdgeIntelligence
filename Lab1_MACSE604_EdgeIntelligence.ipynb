Resizing the images from Training and Testing dataset

````python
from PIL import Image
import os

SRC_TRAIN = "/content/photos50k/train"
DST_TRAIN = "/content/photos50k/train_final"

SRC_TEST  = "/content/photos50k/test"
DST_TEST  = "/content/photos50k/test_final"

SIZE = (224,224)

def resize_tree(src_root, dst_root):
    for root, dirs, files in os.walk(src_root):
        for f in files:
            if f.lower().endswith((".jpg",".jpeg",".png")):
                src_path = os.path.join(root, f)
                rel = os.path.relpath(src_path, src_root)
                dst_path = os.path.join(dst_root, rel)
                os.makedirs(os.path.dirname(dst_path), exist_ok=True)
                im = Image.open(src_path).convert("RGB")
                im = im.resize(SIZE)
                im.save(dst_path)

os.makedirs(DST_TRAIN, exist_ok=True)
os.makedirs(DST_TEST,  exist_ok=True)

resize_tree(SRC_TRAIN, DST_TRAIN)
resize_tree(SRC_TEST,  DST_TEST)
````

Classes changing for test dataset

````python
import os, random, shutil
from pathlib import Path

ROOT = Path("/content/photos50k")
TRAIN = ROOT / "train_final"   #  resized train folder
TEST  = ROOT / "test_final"    # resized test folder
SAMPLE_PER_CLASS = 50          # how many images to ensure in test per class (adjust if you want more)
SEED = 42

random.seed(SEED)

# ensure directories exist
assert TRAIN.exists(), f"train folder missing: {TRAIN}"
TEST.mkdir(parents=True, exist_ok=True)

train_classes = sorted([d.name for d in TRAIN.iterdir() if d.is_dir()])
test_classes  = sorted([d.name for d in TEST.iterdir()  if d.is_dir()])

print("Train classes:", len(train_classes))
print("Test classes :", len(test_classes))

# For every class in train, ensure test has that folder and at least SAMPLE_PER_CLASS images.
for cls in train_classes:
    src_dir = TRAIN / cls
    dst_dir = TEST / cls
    dst_dir.mkdir(parents=True, exist_ok=True)

    # list existing images in dst and src
    src_imgs = [p for p in src_dir.iterdir() if p.suffix.lower() in (".jpg",".jpeg",".png")]
    dst_imgs = [p for p in dst_dir.iterdir() if p.suffix.lower() in (".jpg",".jpeg",".png")]

    # if test already has >= SAMPLE_PER_CLASS, skip
    if len(dst_imgs) >= SAMPLE_PER_CLASS:
        continue

    # choose candidates from train that are not already in test (by filename)
    dst_names = {p.name for p in dst_imgs}
    candidates = [p for p in src_imgs if p.name not in dst_names]

    # if no candidates, skip
    if not candidates:
        continue

    need = SAMPLE_PER_CLASS - len(dst_imgs)
    if need > len(candidates):
        pick = candidates[:]   # take all available
    else:
        pick = random.sample(candidates, need)

    for p in pick:
        dest = dst_dir / p.name
        # copy file (avoid overwrite)
        if not dest.exists():
            shutil.copy2(p, dest)

# Final verification
new_test_classes = sorted([d.name for d in TEST.iterdir() if d.is_dir()])
print("After fix  Train classes:", len(train_classes), "Test classes:", len(new_test_classes))

# Show counts for a few classes
from collections import Counter
cnt = Counter()
for cls in train_classes[:10]:
    cnt[cls] = len(list((TEST/cls).glob("*.*")))
print("Sample test counts (first 10 classes):")
for k,v in cnt.items():
    print(k, v)
````

Reloading the resized images

````python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

transform = transforms.ToTensor()

train_ds = datasets.ImageFolder("/content/photos50k/train_final", transform=transform)
test_ds  = datasets.ImageFolder("/content/photos50k/test_final",  transform=transform)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=64, shuffle=False)

images, labels = next(iter(train_loader))
print("Batch:", images.shape, labels.shape)
print("Train classes:", train_ds.classes)
print("Test  classes:", test_ds.classes)
````
